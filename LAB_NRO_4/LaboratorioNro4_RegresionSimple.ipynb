# Linear Regression

```python
# Importar los módulos necesarios para el desarrollo un modelo de regresión lineal.

%matplotlib notebook
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import os
```

```python
# Generador ramdomico inicializado
rng = np.random.default_rng()

# Generador de los datos ramdomicos
x = rng.random(16)
y = 1.6*x + rng.random(16)
y
```

```python
# Calcula la regresión lineal de mínimos cuadrados para dos conjuntos de datos.
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
# Utilizar la función auxiliar que usa la 
# pendiente e intersección calculadas para 
# devolver un nuevo valor.
def fitted_value(x):
    return slope * x + intercept
```

```python
# Calcular los valores ajustados para el eje ‘y’ usando una función auxiliar
new_values = list(map(fitted_value, x))
```

```python
# Visualización del modelo.
plt.plot(x, y, 'o', label='original data')

# Trazo de la linea de la regresion lineal
plt.plot(x, new_values, 'r', label='fitted line')

plt.legend()


```
